```python
#Colunas selecionadas


colunasAnalise = ['NU_INSCRICAO', 'SG_UF_RESIDENCIA',  'NU_IDADE', 'TP_SEXO', 'TP_ESTADO_CIVIL', 'TP_COR_RACA'
                  ,'IN_TREINEIRO','TP_ANO_CONCLUIU','TP_ESCOLA',
                  'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO','Q024', 'Q025']


```

```python
#Importando os microdados do enem para um dataframe

df = pd.read_csv('MICRODADOS_ENEM_2018.csv', sep = ';', encoding='ISO 8859-1', nrows = 10)

#Ver os valores das coluna do dataframe gerado
colunas = df.columns.values
colunas
```

```python
#Importando os microdados, mas apenas das colunas que ser√£o usadas na analise
# usecols usada pra definir as colunas que ser√£o importadas
# encoding pra dizer o tipo de codifica√ß√£o da base
df_microdadosEnem = pd.read_csv('MICRODADOS_ENEM_2018.csv', sep = ';', encoding='ISO 8859-1', usecols = colunasAnalise)

# Remover as linhas com NAN usando a fun√ß√£o dropna()
df_microdadosEnem.dropna(inplace =True) # o implace = True faz persistir a mudan√ßa na mem√≥ria
df_microdadosEnem.head()
```

```python
# Salvar dataframe com os microdados do enem com as colunas que ser√£o usadas na analise
df_microdadosEnem.to_csv('microdadosEnemSelecionados.csv')
```

```python
# Data frame com os dados de Alagoas e sem os NAN

#fun√ß√£o loc usada pra selecionar as linhas que estiverem de acordo com a retri√ß√£o
df_MicrodadosEnem_Alagoas = df_microdadosEnem.loc[(df_microdadosEnem['SG_UF_RESIDENCIA'] == 'AL')]

#func√£o sample() usada pra mostrar um trecho do dataframe
df_MicrodadosEnem_Alagoas.sample(10)
```

# Tratar os dados, transformar tudo pra o mesmo tipo


```python
df_MicrodadosEnem_Alagoas.TP_SEXO.values
df_MicrodadosEnem_Alagoas.TP_SEXO
```

```python
# tranformar o sexo: 0 = M, 1 = F
from collections import defaultdict

dictSexo = {'M': 0, 'F':1 }
for e in dictSexo:
    print(dictSexo[e])  

df_MicrodadosEnem_Alagoas.TP_SEXO = [ dictSexo [resp] for resp in df_MicrodadosEnem_Alagoas.TP_SEXO]

```

```python
testeQ024 = df_MicrodadosEnem_Alagoas.Q024.str.get_dummies()
testeQ024
df_MicrodadosEnem_AlagoasQ024 = pd.concat([df_MicrodadosEnem_Alagoas, testeQ024], axis = 1) #o atributo axis=1 diz que √© pra concatenar lado a lado, se for axis=0, seria um abaixo do outro

testeQ025 = df_MicrodadosEnem_Alagoas.Q025.str.get_dummies()
testeQ025
df_MicrodadosEnem_AlagoasQ025 = pd.concat([df_MicrodadosEnem_Alagoas, testeQ025], axis = 1)
#df_MicrodadosEnem_Alagoas['Q024'] = 

#df_MicrodadosEnem_Alagoas['Q024'] = 
```

```python
#Salvar num arquivo csv o dataframe com os dados de Alagoas
df_MicrodadosEnem_Alagoas.to_csv('microdadosEnemAlagoas.csv')
```
