```python
#Colunas selecionadas


colunasAnalise = ['NU_INSCRICAO', 'SG_UF_RESIDENCIA',  'NU_IDADE', 'TP_SEXO', 'TP_ESTADO_CIVIL', 'TP_COR_RACA',
                  'IN_TREINEIRO','TP_ANO_CONCLUIU','TP_ESCOLA','NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 
                  'NU_NOTA_MT', 'NU_NOTA_REDACAO',                  
                  'Q001', 'Q002', 'Q003', 'Q004', 'Q005', 'Q006','Q007', 'Q008', 'Q009', 'Q010', 'Q011', 'Q012',
                  'Q013', 'Q014', 'Q015', 'Q016', 'Q017', 'Q018', 'Q019', 'Q020', 'Q021', 'Q022','Q023', 'Q024',
                  'Q025', 'Q026', 'Q027']



```

```python
#Importando os microdados do enem para um dataframe

df = pd.read_csv('MICRODADOS_ENEM_2018.csv', sep = ';', encoding='ISO 8859-1', nrows = 10)

#Ver os valores das coluna do dataframe gerado
#colunas = df.columns.values

#colunas 
```

```python
#Importando os microdados, mas apenas das colunas que serão usadas na analise
# usecols usada pra definir as colunas que serão importadas
# encoding pra dizer o tipo de codificação da base
df_microdadosEnem = pd.read_csv('MICRODADOS_ENEM_2018.csv', sep = ';', encoding='ISO 8859-1', usecols = colunasAnalise)

# Remover as linhas com NAN usando a função dropna()
df_microdadosEnem.dropna(inplace =True) # o implace = True faz persistir a mudança na memória
df_microdadosEnem.head()
```

```python
# Salvar dataframe com os microdados do enem com as colunas que serão usadas na analise
df_microdadosEnem.to_csv('microdadosEnemSelecionados.csv', sep = ';', encoding='ISO 8859-1')
```

```python
# Data frame com os dados de Alagoas e sem os NAN

#função loc usada pra selecionar as linhas que estiverem de acordo com a retrição
df_MicrodadosEnem_Alagoas = df_microdadosEnem.loc[(df_microdadosEnem['SG_UF_RESIDENCIA'] == 'AL')]

#funcão sample() usada pra mostrar um trecho do dataframe
df_MicrodadosEnem_Alagoas.sample(10)
```

```python
#Salvar num arquivo csv o dataframe com os dados de Alagoas
df_MicrodadosEnem_Alagoas.to_csv('microdadosEnemAlagoas.csv', sep = ';', encoding='ISO 8859-1')
```
